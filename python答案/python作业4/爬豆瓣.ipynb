```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
from lxml import etree
headers={"User-Agent":"Mozilla/5.0 (Windows NT 6.1; WOW64; rv:61.0) Gecko/20100101 Firefox/61.0"}
r=requests.get("https://movie.douban.com/top250")
bsObj=BeautifulSoup(re.sub(" +"," ",r.text))
print(bsObj)
print(type(bsObj))
```

```python
nodes=bsObj.find_all("div",{'class':'item'})
```

```python
node=nodes[-1]
```

```python
node
```

```python
rank=node.find('em',{'class':''})
rank
```

```python
name=node.find('span',{'class':'title'})
name
```

```python
basic_info=node.find('p',{'class':''})
basic_info
```

```python
score=node.find('span',{'class':'rating_num'})
score
```

```python
comment=node.find('span',{'class':'inq'})
comment
```

```python
info=[]
info.append([name.text,basic_info.text,score.text,comment.text])
```

```python
info=[]
for node in nodes:
    rank=node.find('em',{'class':''}).text
    name=node.find('span',{'class':'title'}).text
    basic_info=node.find('p',{'class':''}).text
    score=node.find('span',{'class':'rating_num'}).text
    comment=node.find('span',{'class':'inq'}).text
    info.append([rank,name,basic_info,score,comment])
```

```python
df=pd.DataFrame(info)
```

```python
df
```

```python
df.columns=['排名','影名','基本信息','评分','简评']
```

```python
df
```

```python
df=df.set_index('排名')
```

```python
df
```

```python
df['基本信息']=df['基本信息'].map(lambda x:x.replace('\n',' '))
```
